```python
## renaming columns
## Renaming columns
def rename_columns(data: pd.DataFrame):
  return (
      data.rename(str.lower,axis = 1)
      .rename(columns = {
          "delivery_person_id" : "rider_id",
          "delivery_person_age" : "age",
          "delivery_person_ratings" :"ratings",
          "delivery_location_latitude" : "delivery_latitude",
          "delivery_location_longitude" : "delivery_longitude",
          "time_ordered" : "order_time",
          "time_order_picked" : "order_picked_time",
          "weatherconditions" : "weather",
          "road_traffic_density" : "traffic",
          "time_taken(min)" : "time_taken"
      })
  )
```

**weather**


* 616 records are present in weather having missing values.

```python
def clean_weather(data: pd.DataFrame):
  return(data.assign(
      weather = lambda x:x["weather"].str.replace("conditions","").str.lower().str.strip().replace("nan",np.nan)
  ))
```

**time_taken**

```python
df["time_taken"].info()
```

> time_taken is present in string format, need to changed to int. <br>
> values contains (min) , should be removed.

```python
def clean_target(data: pd.DataFrame):
  return (data.assign(
      time_taken = lambda x: x["time_taken"].str.replace("(min)","").astype(int)
  ))
```

```python
clean_target(df).info()
```

> Using pipe to integrate whole process of cleaning.

```python
cleaned_data = (
    df.pipe(drop_id)
    .pipe(split_rider_id)
    .pipe(clean_age,minor_data_index)
    .pipe(clean_ratings,six_star_index)
    .pipe(clean_restaurant_latitude)
    .pipe(clean_restaurant_longitude)
    .pipe(clean_order_date)
    .pipe(clean_order_time)
    .pipe(clean_order_picked_time)
    .pipe(clean_weather)
    .pipe(clean_traffic)
    .pipe(clean_festival)
    .pipe(clean_type_of_order)
    .pipe(clean_type_of_vehicle)
    .pipe(clean_city)
    .pipe(clean_target)
)
```
