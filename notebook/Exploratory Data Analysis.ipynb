## Feature Engineering

```python
df = data.copy()
```

```python
def calculate_haversine_distance(data: pd.DataFrame):
    # Convert degrees to radians
    lat1 = np.radians(df["delivery_latitude"])
    lon1 = np.radians(df["delivery_longitude"])
    lat2 = np.radians(df["restaurant_latitude"])
    lon2 = np.radians(df["restaurant_longitude"])

    # Differences
    dlat = lat2 - lat1
    dlon = lon2 - lon1

    # Haversine formula
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    distance_km = 6371 * c  # Earth's radius in kilometers

    df["distance(km)"] = distance_km
    df['distance(km)'] = df['distance(km)'].round(2)

    return df
```

**As the data belongs to the same year, so year do not play any significant role in delivery time prediction.**

```python
def extract_date_features(data: pd.DataFrame):
  return(data.assign(
      order_date = lambda x: pd.to_datetime(x['order_date'],dayfirst=True),
      order_day = lambda x: x['order_date'].dt.day,
      order_month = lambda x: x['order_date'].dt.month,
      order_day_of_week = lambda x: x['order_date'].dt.day_name().str.lower(),
      is_weekend = lambda x: (x['order_date']
                              .dt.day_name()
                              .isin(["Saturday","Sunday"])
                              .astype(int))
      )
  )
```

```python
def extract_time_features(data: pd.DataFrame) -> pd.DataFrame:
    # Safely convert to datetime
    data['order_time'] = pd.to_datetime(data['order_time'], errors='coerce', format='mixed')
    data['order_picked_time'] = pd.to_datetime(data['order_picked_time'], errors='coerce', format='mixed')

    # Calculate pickup time only if timestamps are valid and in correct order
    data['pickup_time_minutes'] = np.where(
        data['order_time'].notna() & data['order_picked_time'].notna() &
        (data['order_picked_time'] >= data['order_time']),
        (data['order_picked_time'] - data['order_time']).dt.total_seconds() / 60,
        np.nan  # or set to 0 or flag as invalid
    )

    # Extract hour from order_time
    data['order_hour'] = data['order_time'].dt.hour

    # Define time_of_day logic
    conditions = [
        (data['order_hour'] >= 6) & (data['order_hour'] < 12),
        (data['order_hour'] >= 12) & (data['order_hour'] < 17),
        (data['order_hour'] >= 17) & (data['order_hour'] < 20),
        (data['order_hour'] >= 20) & (data['order_hour'] < 24)
    ]
    labels = ["morning", "afternoon", "evening", "night"]

    # Apply time_of_day categorization
    data['order_period'] = np.select(conditions, labels, default="after_midnight")

    return data
```

```python
def remove_columns(data: pd.DataFrame):
  return(data.drop(
      columns = ["order_date","order_time","order_picked_time"]
  )
  )
```

```python
transformed_data = (
    data.pipe(calculate_haversine_distance)
    .pipe(extract_date_features)
    .pipe(extract_time_features)
    .pipe(remove_columns)
)
```

### Visualization & Hypothesis testing

```python
def numerical_analysis(dataframe, column_name, cat_col=None, bins="auto"):
    # Create the figure with a specified size
    fig = plt.figure(figsize=(15, 10))

    # Set up the subplot layout using add_subplot
    ax1 = fig.add_subplot(2, 2, 1)  # Top-left plot
    ax2 = fig.add_subplot(2, 2, 2)  # Top-right plot
    ax3 = fig.add_subplot(2, 1, 2)  # Bottom plot spanning both columns

    # Plot the KDE Plot
    sns.kdeplot(data=dataframe, x=column_name, hue=cat_col, ax=ax1)
    ax1.set_title(f'KDE Plot of {column_name}')

    # Plot the Box Plot
    sns.boxplot(data=dataframe, x=column_name, hue=cat_col, ax=ax2)
    ax2.set_title(f'Box Plot of {column_name}')

    # Plot the Histogram
    sns.histplot(data=dataframe, x=column_name, bins=bins, hue=cat_col, kde=True, ax=ax3)
    ax3.set_title(f'Histogram of {column_name}')

    plt.tight_layout()
    plt.show()
```

```python
def categorical_analysis(dataframe, column_name):
    print(f"Analysis for the column: '{column_name}'")
    print("-" * 50)

    # Create and display the DataFrame with counts and percentages
    value_counts_df = pd.DataFrame({
        "Count": dataframe[column_name].value_counts(),
        "Percentage": dataframe[column_name].value_counts(normalize=True).mul(100).round(2).astype(str) + "%"
    })
    display(value_counts_df)

    print("-" * 50)

    # Get and print unique categories and their count
    unique_categories = dataframe[column_name].unique().tolist()
    number_of_categories = dataframe[column_name].nunique()

    print(f"The number of unique categories is: {number_of_categories}")
    print(f"The unique categories are: {unique_categories}")

    print("-" * 50)

    # Plot the countplot
    plt.figure(figsize=(10, 6))
    sns.countplot(data=dataframe, x=column_name)
    plt.xticks(rotation=45, ha='right')
    plt.title(f'Count of Each Category in {column_name}')
    plt.tight_layout()
    plt.show()
```

```python
def numerical_categorical_analysis(dataframe, cat_column, num_column):
    # Create a 2x2 grid of subplots
    fig, ax = plt.subplots(2, 2, figsize=(15, 7.5))

    # plot the barplot
    sns.barplot(data=dataframe, x=cat_column, y=num_column, ax=ax[0, 0])

    # plot the boxplot
    sns.boxplot(data=dataframe, x=cat_column, y=num_column, ax=ax[0, 1])

    # plot the violin plot
    sns.violinplot(data=dataframe, x=cat_column, y=num_column, ax=ax[1, 0])

    # plot the strip plot
    sns.stripplot(data=dataframe, x=cat_column, y=num_column, ax=ax[1, 1])

    plt.tight_layout()
    plt.show()
```

```python
def multivariate_analysis(df, num_col, cat_col1, cat_col2):
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 7.5))
    plot_params = dict(data=df, x=cat_col1, y=num_col, hue=cat_col2)

    # Barplot
    sns.barplot(**plot_params, ax=axes[0, 0])
    axes[0, 0].set_title("Barplot")

    # Boxplot
    sns.boxplot(**plot_params, ax=axes[0, 1])
    axes[0, 1].set_title("Boxplot")

    # Violinplot
    sns.violinplot(**plot_params, ax=axes[1, 0])
    axes[1, 0].set_title("Violinplot")

    # Stripplot
    sns.stripplot(**plot_params, dodge=True, ax=axes[1, 1])
    axes[1, 1].set_title("Stripplot")

    plt.tight_layout()
    plt.show()
```

```python
from scipy.stats import chi2_contingency, f_oneway, jarque_bera, probplot
```

```python
def chi_2_test(dataframe, col1, col2, alpha= 0.05):
    data = (
        dataframe.loc[:, [col1, col2]]
        .dropna()
    )
    # create contingency table
    contingency_table = pd.crosstab(data[col1], data[col2])
    # perform chi-squared test
    _, p_val, _, _ = chi2_contingency(contingency_table)
    print(p_val)
    if p_val <= alpha:
        print(f"Reject the null hypothesis. There is a significant association between {col1} and {col2}.")
    else:
        print(f"Fail to reject the null hypothesis. There is no significant association between {col1} and {col2}.")
```

```python
def anova_test(dataframe, num_col, cat_col, alpha=0.05):
    data = (
        dataframe.loc[:, [num_col, cat_col]]
        .dropna()
    )
    cat_group = data.groupby(cat_col)
    groups = [group[num_col].values for _, group in cat_group]
    f_stat, p_val = f_oneway(*groups)
    print(p_val)
    if p_val <= alpha:
        print(f"Reject the null hypothesis. There is a significant relationship between {num_col} and {cat_col}.")
    else:
        print(f"Fail to reject the null hypothesis. There is no significant relationship between {num_col} and {cat_col}.")
```

```python
def test_for_normality(dataframe, column_name, alpha=0.05):
    data = dataframe[column_name]
    print("Jarque Bera Test for Normality")
    _, p_val = jarque_bera(data)
    print(p_val)
    if p_val <= alpha:
        print(f"Reject the null hypothesis. The data is not normally distributed.")
    else:
        print(f"Fail to reject the null hypothesis. The data is normally distributed.",end="\n\n")
```
